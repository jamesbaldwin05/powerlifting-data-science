```python
import numpy as np
import pandas as pd

fulldata = pd.read_csv('data/openipf-2025-11-08-c1c550e2.csv')
```

```python
data = fulldata[['Name', 'Sex', 'Age', 'BodyweightKg', 'Best3SquatKg', 'Best3BenchKg', 'Best3DeadliftKg', 'Equipment']]
data = data[data['Equipment'] == 'Raw']
data = data.dropna()
data = data.drop(columns=['Equipment'])
```

```python
data.drop_duplicates(subset=['Name'], keep=False,inplace=True)
```

```python
data = data[data['BodyweightKg'] < 93 ]
data = data[data['Sex']=='M']
```

```python
data = data[data['BodyweightKg'] > 83 ]
data = data[(data['Age'] >= 20) & (data['Age'] <= 35)]
```

```python
squatdata = data.sort_values(by='Best3SquatKg', ascending=False)
squat1 = squatdata[squatdata['Best3SquatKg'] > 250]
squat2 = squatdata[squatdata['Best3SquatKg'] > 260]
squat3 = squatdata[squatdata['Best3SquatKg'] > 280]
squatbase = squatdata[squatdata['Best3SquatKg'] > 180]
print(f"Percentage of lifters with squat > 250kg: {len(squat1)/len(squatbase)*100:.2f}%")
print(f"Percentage of lifters with squat > 260kg: {len(squat2)/len(squatbase)*100:.2f}%")
print(f"Percentage of lifters with squat > 280kg: {len(squat3)/len(squatbase)*100:.2f}%")
```

```python
benchdata = data.sort_values(by='Best3BenchKg', ascending=False)
bench1 = benchdata[benchdata['Best3BenchKg'] > 150]
bench2 = benchdata[benchdata['Best3BenchKg'] > 160]
bench3 = benchdata[benchdata['Best3BenchKg'] > 180]
benchbase = benchdata[benchdata['Best3BenchKg'] > 100]
print(f"Percentage of lifters with bench > 150kg: {len(bench1)/len(benchbase)*100:.2f}%")
print(f"Percentage of lifters with bench > 160kg: {len(bench2)/len(benchbase)*100:.2f}%")
print(f"Percentage of lifters with bench > 180kg: {len(bench3)/len(benchbase)*100:.2f}%")
```

```python
deadliftdata = data.sort_values(by='Best3DeadliftKg', ascending=False)
deadlift1 = deadliftdata[deadliftdata['Best3DeadliftKg'] > 280]
deadlift2 = deadliftdata[deadliftdata['Best3DeadliftKg'] > 300]
deadlift3 = deadliftdata[deadliftdata['Best3DeadliftKg'] > 320]
deadliftbase = deadliftdata[deadliftdata['Best3DeadliftKg'] > 200]
print(f"Percentage of lifters with deadlift > 280kg: {len(deadlift1)/len(deadliftbase)*100:.2f}%")
print(f"Percentage of lifters with deadlift > 300kg: {len(deadlift2)/len(deadliftbase)*100:.2f}%")
print(f"Percentage of lifters with deadlift > 320kg: {len(deadlift3)/len(deadliftbase)*100:.2f}%")
```
