```python
import numpy as np
import pandas as pd

fulldata = pd.read_csv('../data/openipf-2025-11-08-c1c550e2.csv')
```

```python
# Keep only unique names (so we don't double count lifters with multiple meets), I know this algorithm is inefficient but its a smallish dataset
unique = fulldata.copy()
unique = unique[unique['Event']=='SBD']
unique = unique[unique['Tested']=='Yes']
unique.sort_values(['Name', 'Event', 'Equipment', 'WeightClassKg', 'TotalKg'], ascending=True, inplace=True)
```

```python
unique = unique.drop_duplicates(
    subset=['Name', 'Event', 'Equipment', 'WeightClassKg'],
    keep='last'
)
```

```python
data = unique[['Sex', 'Equipment', 'AgeClass', 'WeightClassKg', 'Best3SquatKg', 'Best3BenchKg', 'Best3DeadliftKg', 'TotalKg']].copy()
```

```python
userData = ['M', 'Raw', '20-23', '93', 182.5, 100, 200, 482.5]
```

```python
filtered_data = data[data['Sex'] == userData[0]]
filtered_data1 = filtered_data.loc[filtered_data['Equipment'] == userData[1]]
filtered_data2 = filtered_data1.loc[filtered_data1['AgeClass'] == userData[2]]
filtered_data3 = filtered_data2.loc[filtered_data2['WeightClassKg'] == userData[3]]
```

```python
subdata = filtered_data3

squat_percentile = (subdata['Best3SquatKg'] < userData[4]).sum() / len(subdata) * 100
bench_percentile = (subdata['Best3BenchKg'] < userData[5]).sum() / len(subdata) * 100
deadlift_percentile = (subdata['Best3DeadliftKg'] < userData[6]).sum() / len(subdata) * 100
total_percentile = (subdata['TotalKg'] < userData[7]).sum() / len(subdata) * 100
```

```python
print(f"Squat Percentile: {squat_percentile:.2f}")
print(f"Bench Percentile: {bench_percentile:.2f}")
print(f"Deadlift Percentile: {deadlift_percentile:.2f}")
print(f"Total Percentile: {total_percentile:.2f}")
```
