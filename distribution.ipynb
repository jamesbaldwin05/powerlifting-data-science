```python
import numpy as np
import pandas as pd

fulldata = pd.read_csv('data/openipf-2025-11-08-c1c550e2.csv')

data = fulldata[['Sex', 'Age', 'BodyweightKg', 'Best3SquatKg', 'Best3BenchKg', 'Best3DeadliftKg', 'Equipment']]
data = data[data['Equipment'] == 'Raw']
data = data.dropna()
data = data.drop(columns=['Equipment'])
```

```python
data = data[data['BodyweightKg'] < 93 ]
data = data[data['Sex']=='M']
```

```python
data = data[data['BodyweightKg'] > 83 ]
data = data[(data['Age'] >= 20) & (data['Age'] <= 35)]
```

```python
squatdata = data.sort_values(by='Best3SquatKg', ascending=False)
benchdata = data.sort_values(by='Best3BenchKg', ascending=False)
deadliftdata = data.sort_values(by='Best3DeadliftKg', ascending=False)
```

```python
from matplotlib import pyplot as plt

plt.hist(squatdata['Best3SquatKg'], bins=30, alpha=0.5, label='Squat')
plt.xlabel('Best 3 Squat (kg)')
plt.ylabel('Number of Lifters')
plt.xlim(60, 400)
plt.title('Distribution of Best 3 Squat')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.hist(benchdata['Best3BenchKg'], bins=30, alpha=0.5, label='Bench')
plt.xlabel('Best 3 Bench (kg)')
plt.ylabel('Number of Lifters')
plt.xlim(50, 250)
plt.title('Distribution of Best 3 Bench')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.hist(deadliftdata['Best3DeadliftKg'], bins=30, alpha=0.5, label='Deadlift')
plt.xlabel('Best 3 Deadlift (kg)')
plt.ylabel('Number of Lifters')
plt.xlim(60, 400)
plt.title('Distribution of Best 3 Deadlift')
plt.legend()
plt.grid()
plt.show()
```
